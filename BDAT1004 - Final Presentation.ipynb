```python
import pandas as pd
import seaborn as sb
Hospital = pd.read_csv("Hospital_data.csv")
print(Hospital.head())
```

```python
Hospital.shape
```

```python
Hospital_clean = Hospital.dropna()
Hospital_clean
```

```python
Hospital_clean.groupby("Department").Admission_Deposit.sum().sort_values(ascending=False).plot.bar(ylabel = "Admissions Deposit ($ Billions)", title = "Total Admissions Deposit Per Department in 2021")
```

```python
Hospital_clean.groupby("Department").Department.count().sort_values(ascending=False).plot.pie(ylabel = "# of Cases", title = "Cases Per Department in 2021")
```

```python
Hospital_clean.groupby("Department").Admission_Deposit.sum().sort_values(ascending=False).plot.bar(ylabel = "Admissions Deposit ($ Billions)", title = "Total Admissions Deposit Per Department in 2021")
```

```python
Hospital_clean.groupby("Hospital_code").Admission_Deposit.sum().sort_values(ascending=False).plot.bar(xlabel = "Hospital ID", ylabel = "Admissions Deposit ($ Billions)", title = "Total Admissions Deposit Per Hospital")
```

```python
Hospital_clean.groupby("Hospital_code").Hospital_code.count().sort_values(ascending=False)[:5].plot.bar(xlabel = "Hospital ID", ylabel = "# of Cases", title = "Top 5 Hospitals by # Of Cases Per Hospital")
```

```python
Extreme_cases = Hospital_clean[(Hospital_clean['Severity of Illness'] == 'Extreme') | (Hospital_clean['Severity of Illness'] == 'Moderate')]
Extreme_cases
```

```python
Extreme_cases.groupby("Hospital_code").Hospital_code.count().sort_values(ascending=False).plot.bar(xlabel = "Hospital ID", ylabel = "# of Cases (Extreme or Moderate Illness)", title = "Hospitals ranked by # Of Cases Of Extreme or Moderate Illnesses")
```
